```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression
import sklearn 
import pandas as pd
import numpy as np
import warnings as wr
wr.filterwarnings('ignore')
#import all the necessary library 
```

```python
df= pd.read_csv('HW.csv') #read file
df.shape
```

```python
df.Height.mean() #avg of height
df.Weight.mean() #avg of weight
x=df.iloc[:,0] # sent height column into x value
y=df.iloc[:,1] #sent weight column into y value
```

```python
xtrain, xtest, ytrain, ytest = tts(x,y,test_size=0.33, random_state=1) #train model
```

```python
model= LinearRegression()
x_train= np.array(xtrain).reshape(-1,1) #making 2d 
y_train= np.array(ytrain).reshape(-1,1)
x_test =np.array(xtest).reshape(-1,1)

model.fit(x_train,y_train)

model.predict([[2000]]) # using tained model
```

```python
m = model.coef_
c = model.intercept_
predicted = (m*2000) + c # using formula 
predicted
```

```python
plt.plot(xtest, model.predict(x_test), color='red')
plt.scatter(df[['Height']] , df[['Weight']], color='green', marker='+')
```
